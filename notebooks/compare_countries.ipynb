Statistical Testing

```python
from scipy import stats
```

```python
# Extract GHI values
ghi_benin = df[df["Country"] == "Benin"]["GHI"]
ghi_sirralone = df[df["Country"] == "Sirralone"]["GHI"]
ghi_togo = df[df["Country"] == "Togo"]["GHI"]
```

one-way ANOVA Test

```python
f_stat,p_value = stats.f_oneway(ghi_benin,ghi_sirralone,ghi_togo) # find f_stat and p_value conduct using one-way ANOVA method for GHI acros three countries

print("f_stat:",f_stat)
print(f"p_value:{p_value:.10f}")

```

p_value report

A one-way ANOVA was conducted to assess whether there are statistically significant differences in Global Horizontal Irradiance (GHI) values among Benin, Sierra Leone, and Togo. The analysis included hundreds of thousands of observations per country (Benin: 454,311; Sierra Leone: 402,476; Togo: 512,281), and the test revealed a highly significant result (F = 2746.36, p < 0.001). This extremely low p-value is primarily due to the very large sample size, which allows even modest differences in mean GHI to become statistically detectable. Examination of the summary statistics shows that the mean GHI values are relatively close (Benin: 274.76, Sierra Leone: 258.45, Togo: 227.18) with substantial within-country variability, indicating that while the differences are statistically significant, their practical impact on solar radiation patterns is limited. Therefore, stakeholders should interpret these results considering both statistical significance and the actual magnitude of differences across countries.

KruskalWallis Test

```python
f_stat,p_value = stats.kruskal(ghi_benin,ghi_sirralone,ghi_togo) # find f_stat and p_value conduct using KruskalWallis method for GHI acros three countries

print("f_stat:",f_stat)
print(f"p_value:{p_value:.10f}")
```

p_value report

A KruskalWallis H-test was conducted to assess differences in Global Horizontal Irradiance (GHI) among Benin, Sierra Leone, and Togo. The test included several hundred thousand observations per country and yielded an extremely high H-statistic (H = 9994.92) with a p-value effectively equal to zero (p < 0.001). This indicates a statistically significant difference in the distribution of GHI values between at least one pair of countries. While the p-value is extremely small due to the very large sample size, examination of the median values (Benin: 64.2, Sierra Leone: 111.7, Togo: 1.7) shows that Togo has a notably lower typical GHI compared to the other countries. Therefore, the differences in solar radiation across these countries are not only statistically significant but also have practical implications for applications such as solar energy planning.

Key Observations

1 Benin has the highest mean GHI (274.76), but it also shows very high variability (std = 337.86), indicating wide fluctuations in solar irradiance.

2 Sierra Leone has a slightly lower mean GHI (258.45) but a much higher median (111.7), suggesting a skewed distribution with some very high values.

3 Togo shows the lowest median GHI (1.7) despite a moderate mean (227.18), highlighting a distribution with many very low values and high variability.
